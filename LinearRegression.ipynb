```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
```

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
from sklearn.linear_model import LinearRegression 
rg = LinearRegression()
```

```python
rg.fit(x_train, y_train)
```

```python
x_pred = rg.predict(x_train)
y_pred = rg.predict(x_test)
```

```python
print("Data Latih")
pt.scatter(x_train, y_train, color="red")
pt.plot(x_train, x_pred, color="green")
pt.show()
```

```python
print("data test")
pt.scatter(x_test, y_test, color="blue")
pt.plot(x_test, y_pred, color="yellow")
pt.show()
```
